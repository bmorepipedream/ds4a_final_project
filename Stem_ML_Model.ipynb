```python
import pandas as pd
import numpy as np
import csv
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats

#Import each Dataset as a DF.
df=pd.read_csv('stem_final_2.csv')

# Multivariate Linear Regression formula
f5= 'income ~ yearsofexperience + yearsatcompany + title + level + education +  gender + state + race + hispanic'
#fitted model
model5 = smf.ols(formula = f5, data = df)
#fit Model
fitted5 = model5.fit()
```

```python
fitted5.summary()
```

```python
#save model as pickle object
import pickle    
with open('model.pkl','wb') as f:
  pickle.dump(fitted5,f)
```
